```python
import pandas as pd
import numpy as np
from glob import glob
```

```python
files = glob('./*.txt')
```

```python
for file in files:
    acc = np.zeros(178)
    with open(file) as fp:
        temp = (fp.read()[1:-1].split(','))
        temp = [int(t) for t in temp]
        
        for i, t in enumerate(temp):
            acc[i] += temp[i]
print(acc)
```
